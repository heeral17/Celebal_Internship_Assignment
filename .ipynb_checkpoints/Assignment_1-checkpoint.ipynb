```python
# Importing libraries
import numpy as np 
import pandas as pd
import os
import glob
```

```python
# defining dataset location
folder_path = "/Users/jain/Downloads/archive"
```

```python
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
```

```python
#Concatenating data into one dataset 
df_list = []

for file in csv_files:
    df_list.append(pd.read_csv(file))

combined_df = pd.concat(df_list, ignore_index=True)

print(combined_df)
```

```python
#Checking Columns type of each files
for file in csv_files:
    df = pd.read_csv(file)
    print("Columns:", df.columns)
```

```python
combined_df.fillna({"price":0, "category":"Unknown"})
```

```python
combined_df = combined_df.drop_duplicates()
combined_df
```

```python
filtered_df = combined_df[combined_df["initial_price"] < 2000]
filtered_df
```

```python
selected_df = combined_df[["product_id", "title", "category", "final_price", "discount"]]
selected_df
```

```python
combined_df["quantity"] = 1
combined_df["total_amount"] = combined_df["initial_price"] * 1
```

```python
combined_df.to_csv("cleaned_dataset.csv", index=False)
df=pd.read_csv('cleaned_dataset.csv')
df
```
